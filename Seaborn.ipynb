## Data Visualization with Seaborn

### Seaborn is a high-level statistical data visualization library built on top of Matplotlib. It provides attractive default styles and simplifies complex visualizations with minimal code.

```python
# pip install seaborn
```

```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn

tips = sns.load_dataset("tips")
tips
```

```python
# Create a Scatter Plot

import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
```

```python
# Create a Line Plot

sns.lineplot(x="size", y="total_bill", data=tips, color="red", marker="o")
plt.title("Line Plot of People Vs Total Bill")
plt.xlabel("No of Customers")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Categorical Plots -> non numerical columns
# 1) Bar Plot
# 2) Box Plot
# 3) Violin Plot

sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot of Day vs Total Bill")
plt.show()
```

```python
# Catergorical Plot
# Box Plot

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Day vs Total Bill")
```

```python
# Categorical Plot
# Violin Plot

sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Day vs Total Bill")
```

```python
# Histogram Plot

sns.histplot(tips["total_bill"], bins=10)

sns.histplot(tips["total_bill"], bins=10, kde=True)
```

```python
# KDE (Kernal Density Estimator) Plot

sns.kdeplot(tips["total_bill"])
```

```python
# KDE (Kernel Density Estimator)

sns.kdeplot(tips["total_bill"], fill=True)
```

```python
# Pair Plot -> relationship between each of the numerical columns datas.

sns.pairplot(tips)
```

```python
# Correlation works with only numerical columns datas.

corr = tips[["total_bill", "tip", "size"]].corr()

corr
```

```python
# Heat Map

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heat Map For Correlation of Numerical Columns")
plt.show()
```

```python
# Working on Real Life Datasets

import pandas as pd

df = pd.read_csv("sales_data.csv")

df.head()
```

```python
# Bar Plot of Product Category vs Total Revenue

plt.figure(figsize=(8, 8))
sns.barplot(x="Product Category", y="Total Revenue", data=df, estimator=sum)
plt.title("Bar Plot of Product Category vs Total Revenue")
plt.show()
```

```python
# Bar Plot of Region vs Total Revenue

plt.figure(figsize=(8, 8))
sns.barplot(x="Region", y="Total Revenue", data=df, estimator=sum)
plt.title("Bar Plot of Region vs Total Revenue")
plt.show()
```
